```python
import cv2
from ultralytics import YOLO
from PIL import Image

# Load the trained YOLOv8 model
model = YOLO("yolov8n_trained.pt")

# Open the video source (e.g., webcam)
cap = cv2.VideoCapture(r"C:\Users\ADHISH S\Desktop\Crowd_CV\Datasets\Motion_Emotion_Dataset\007.mp4")

while True:
    # Read a frame from the video source
    ret, frame = cap.read()

    # If the frame was successfully read...
    if ret:
        # Convert the frame to a format that the model can use
        img = Image.fromarray(frame)

        # Run the model on the frame
        results = model.predict(img)

        # Draw the detection results on the frame
        for result in results[0]:
            x1, y1, x2, y2 = result[0:4]
            label = result[-1]
            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

        # Display the frame with the detection results
        cv2.imshow("Crowd Behavior Detection", frame)

        # Exit the loop if the user presses the 'q' key
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release the video source and destroy all windows
cap.release()
cv2.destroyAllWindows()
```

```python
import os

from ultralytics import YOLO
import cv2


VIDEOS_DIR = os.path.join('.', 'videos')

video_path =r"C:\Users\ADHISH S\Desktop\Crowd_CV\Datasets\Motion_Emotion_Dataset\008.mp4"
video_path_out = '{}_out.mp4'.format(video_path)

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))


# Load a model
model = YOLO("yolov8n_trained.pt") # load a custom model

threshold = 0.5

while ret:

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import os
import cv2
from ultralytics import YOLO

VIDEOS_DIR = os.path.join('.', 'videos')

video_path = r"C:\Users\ADHISH S\Desktop\Realtime\video_2.mp4"
video_path_out = '{}_out.mp4'.format(video_path)

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

# Load a model
model = YOLO("best_detect.pt") # load a custom model

threshold = 0.1

while ret:

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    # Display the frame with detections
    cv2.imshow('Real-time Detection', frame)
    
    # Write the frame to the output video
    out.write(frame)
    
    # Check for user input to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
